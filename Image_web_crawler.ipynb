Find (sub)species of animal using above provided name and url

```python
print("Looking for wiki: {}\nStart:\n".format(wiki_page))
source_code = requests.get(wiki_page)
plain_text = source_code.text
soup = BeautifulSoup(plain_text)
class_animal_list = []

# does not work in an automated fashion, manual labour needed

do_not_care_for = ["wiki","pubmed","taxonomy","von","vaan", "cellular", "classified",".","-"]
for link in soup.find_all("a", {"title": {"species","subspecies"}}):
    # "title": "species"
#     "title":"subspecies"
    animal_sub_type = link.string
    animal_sub_type = animal_sub_type.lower()
    found_discard_reason = 0
#     print(animal_sub_type)
    for caution in do_not_care_for:
#         print(animal_sub_type.find(caution))
        if animal_sub_type.find(caution) != -1:
            print("{} - {}".format(animal_sub_type, caution))
            found_discard_reason += 1
    if found_discard_reason == 0:
        class_animal_list.append(animal_sub_type)
            
print("I found {} classes of {}:\n".format(len(class_animal_list), class_animal))
for item in class_animal_list:
    print(item)
```

Download images for each species of previously defined animal and save them to coresponding folder

```python
saveto_folder = "./PetImages/{}/".format(class_animal)

if os.path.isdir(saveto_folder) is False:
    os.mkdir(saveto_folder)
    print("Folder {} created!".format(class_animal))
    
number_of_downloads = 0 # keep tally of downloaded images
download = True

# go trough each species of animal and download images (only first page works yet)
for id,searchword in enumerate(class_animal_list):
    print("[{}/{}]{} - {}".format(id, len(class_animal_list)-1,class_animal, searchword))
    page = 1
    max_pages = 1
    while page <= max_pages:
        url = "https://www.google.com/search?q={}%20{}&tbm=isch&hl=de&tbs=itp:photo,ift:jpg".format(class_animal, searchword, (page*10))
        print("Looking for url: {}".format(url))
        source_code = requests.get(url)
        plain_text = source_code.text
        soup = BeautifulSoup(plain_text)
        for link in soup.find_all("img", {"alt":"Bildergebnis für {} {}".format(class_animal, searchword)}):
            #"class":"rg_ic rg_i",
            # {"class": "rg_1"}
            image_source = link.get("src")
#             print(image_source)
            if download:
                    number_of_downloads += 1
                    rq = urllib3.PoolManager()
                    rq = rq.request("GET",image_source)
                    #print(rq.data)
                    #res = urllib.urlopen(rq)
                    filename = "image_{}_{}_{}.jpg".format(class_animal, searchword, number_of_downloads)
                    fpdf = open(os.path.join(saveto_folder, filename), 'wb')
                    fpdf.write(rq.data)
                    fpdf.close()
        page += 1
    print("Looked through all pages.")
print("looked through all searchwords. Found {} results.".format(number_of_downloads))
```
